```python
t = var('t')
x = var('x')
y = var('y')
fx(t) = cos(t)
fy(t) = sin(t)+5
tmin = 0
tmax = 2*pi
eps = .05
size = 10
def tanAt(f,t,val):
    return diff(f,t)(val)
def allZeroesInRange(func,a,b,tol=0.0001):
    L = []
    try:
        if(func(a)==0):
            L.append(a)
            a+=tol
        if(func(b)==0):
            L.append(b)
            b-=tol
        x0 = find_root(func,a,b)
        L.append(x0)
        L += allZeroesInRange(func,a,x0-tol,tol)       
        L += allZeroesInRange(func,x0+tol,b,tol)       
    except:
        pass
    return L
def minmax(fx,fy,t,tmin,tmax, eps=.05):
    fxp = diff(fx,t)
    fyp = diff(fy,t)
    roots = allZeroesInRange(fxp,tmin,tmax)
    roots += allZeroesInRange(fyp,tmin,tmax)
    xExtrema = [fx(tmin)]
    yExtrema = [fy(tmin)]
    for root in roots:
        xExtrema.append(fx(root))
        yExtrema.append(fy(root))
    xExtrema.append(fx(tmax))
    yExtrema.append(fy(tmax))
    return [(min(xExtrema),min(yExtrema)), (max(xExtrema),max(yExtrema))]
def tangentSpace(fx, fy, tmin, tmax, num, size):
    eps = (tmax-tmin)/num
    ext = minmax(fx,fy,t,tmin,tmax)
    xmin = ext[0][0] - size
    xmax = ext[1][0] + size
    ymin = ext[0][1] - size
    ymax = ext[1][1] + size
    eqs = []
    for i in sxrange(tmin,tmax,eps):
        dy = tanAt(fy,t,i)
        dx = tanAt(fx,t,i)
        if(dx==0):
            eqs.append(line([(fx(i),ymin),(fx(i),ymax)]))
        else:
            eqs.append( implicit_plot(y==(dy/dx)*(x-fx(i))+fy(i), (x,xmin,xmax), (y,ymin,ymax)))
    return sum(eqs)+parametric_plot((fx(t),fy(t)),(t,tmin,tmax), color = 'red')



```
